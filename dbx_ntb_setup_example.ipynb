```python
from databricks.sdk import WorkspaceClient
from datapact.client import DataPactClient
from datapact.config import DataPactConfig
from yaml import safe_load
```

```python
CONFIG_PATH: str = "./datapact/config.yml"
```

```python
with open(CONFIG_PATH, "r", encoding="utf-8") as f:
    raw_cfg = safe_load(f)
cfg = DataPactConfig
```

```python
w = WorkspaceClient()

client = DataPactClient(workspace_client=w)
```

```python
client.run_validation(
    config=cfg,
    job_name="DataPact Test",
    warehouse_name="datapact_warehouse",
    results_table="datapact.results.run_history",
)
```
